```python
import pandas as pd
from geopy.geocoders import Nominatim
data =pd.read_csv('car_crach.csv', encoding='ISO-8859-1')

geolocator = Nominatim(user_agent="geo_locator")

def get_city_and_department(latitude, longitude):
    location = geolocator.reverse((latitude, longitude), language="fr")
    address = location.raw['address']
    city = address.get('city', '')
    department = address.get('state', '')
    print("Ville:", city, "Département:", department)
    return city, department

data['City'], data['Department'] = zip(*data.apply(lambda row: get_city_and_department(row['Latitude'], row['Longitude']), axis=1))
data.to_csv("nouveau_fichier.csv", index=False)
print("Terminé ! Nouveau fichier CSV créé avec les colonnes 'City' et 'Department'.")
```

```python
import pandas as pd
from geopy.geocoders import Nominatim

def get_location(latitude, longitude):
    geolocator = Nominatim(user_agent="geo_locator")
    location = geolocator.reverse((latitude, longitude))
    return location.address if location else "Lieu introuvable"

# Lire le fichier CSV avec pandas
data = pd.read_csv('car_crach.csv', encoding='ISO-8859-1')

# Boucle sur les données pour obtenir les emplacements
for index, row in data.iterrows():
    latitude = row['Latitude']
    longitude = row['Longitude']
    endroit = get_location(latitude, longitude)
    print("-> Endroit trouvé:", endroit)
```
